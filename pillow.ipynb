```python
from PIL import Image
from PIL import Image, ImageFilter
from IPython.display import display
import os # Import the os module to check for file existence
```

```python
image = Image.open('/content/Sumit.jpg')
```

```python
print("\n--- Example of Cropping Images ---")
# To crop an image, you provide a box tuple (left, upper, right, lower).
# These are pixel coordinates. The 'right' and 'lower' coordinates are
# one pixel *beyond* the last pixel to include.
# Let's crop a 100x100 box from the top-left corner.
# Ensure the crop box is within the image dimensions
crop_box = (0, 0, min(100, image.size[0]), min(100, image.size[1]))
cropped_image = image.crop(crop_box)
print(f"Cropped image (top-left 100x100):")
display(cropped_image)
```

```python
resized_image = image.resize((128,128))
resized_image
```

image.png

```python
print("\n--- Example of Rotating and Flipping Images ---")
# Rotate the image by 90 degrees clockwise
rotated_image = image.rotate(90)
print("Image rotated by 90 degrees:")
display(rotated_image)
```

```python
# Flip the image horizontally (left to right)
flipped_image = image.transpose(Image.FLIP_LEFT_RIGHT)
print("Image flipped horizontally:")
display(flipped_image)
```

image.png

```python
print("\n--- Example of Applying Filters ---")
# PIL provides various filters in the ImageFilter module.
# Example: Apply a blur filter
blurred_image = image.filter(ImageFilter.BLUR)
print("Image with BLUR filter applied:")
display(blurred_image)
```

```python
# Example: Apply an edge enhancement filter
edge_enhanced_image = image.filter(ImageFilter.EDGE_ENHANCE)
print("Image with EDGE_ENHANCE filter applied:")
display(edge_enhanced_image)
```

image.png

```python
print("\n--- Example of Converting between different image formats and color modes ---")
# Convert the image to grayscale ('L' mode)
# 'L' stands for Luminance (black and white) - see reference [1]
grayscale_image = image.convert('L')
print("Image converted to Grayscale ('L' mode):")
print(f"Grayscale image mode: {grayscale_image.mode}")
display(grayscale_image)
```

```python
# Convert the image to RGB mode (if it's not already)
# RGB is for True Color - see reference [1]
rgb_image = image.convert('RGB')
print("Image converted to RGB mode:")
print(f"RGB image mode: {rgb_image.mode}")
display(rgb_image)
```
